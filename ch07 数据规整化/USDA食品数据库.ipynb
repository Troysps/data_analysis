```python
#!/usr/bin/env python3
# -*- coding:utf-8 -*-
import pandas as pd
import json

__author__ = 'Troy'

filename = r'C:\Users\Administrator\Desktop\计算机入门学习\支线：Python for Data Analysis\pydata-book-2nd-edition\pydata-book-2nd-edition\datasets\usda_food\database.json'
with open(filename, 'r', encoding='utf-8') as f:
    db = json.load(f)

db[:5]
```

```python
db[0]['description']
columns = ['description', 'group', 'id', 'manufacturer']
info = pd.DataFrame(db, columns=columns)
info.head()
info.duplicated().sum()
pd.value_counts(info.group)[:10]
```

```python
nutrients = []

for i in db:
    nutrient = pd.DataFrame(i['nutrients'])
    nutrient['id'] = i['id']
    nutrients.append(nutrient)

nutrients = pd.concat(nutrients, ignore_index=True)
nutrients.info()
```

```python
nutrients.duplicated().sum()
```

```python
nutrients = nutrients.drop_duplicates()
```

```python
info.info()
col_rename = {'description': 'food', 'group': 'fgroup'}
info = info.rename(columns=col_rename, copy=False)
col_rename = {'description': 'nutrient', 'group': 'nutgroup'}
nutrients = nutrients.rename(columns=col_rename, copy=False)
```

```python
ndata = pd.merge(info, nutrients, on='id', how='outer')
```

```python
ndata.info()
```

```python
%matplotlib inline
result = ndata.groupby(['nutrient', 'fgroup'])['value'].quantile(0.5)
result['Zinc, Zn'].sort_values(ascending=True).plot(kind='barh')
```

```python
by_nutrient = ndata.groupby(['nutgroup', 'nutrient'])
```

```python
# 各营养成分最为丰富的食物是什么
get_maximum = lambda x: x.xs(x.value.idxmax()) # 可以返回数组中最大（小）值的索引值
get_minimum = lambda x: x.xs(x.value.idxmin())
```

```python
by_nutrient.apply(get_maximum)[['value', 'food']]
```
