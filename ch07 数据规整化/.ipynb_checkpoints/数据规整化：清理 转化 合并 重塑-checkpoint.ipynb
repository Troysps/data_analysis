# 数据规整化：清理、转化、合并、重塑

## 合并数据集

## pandas对象中的数据可以通过一些内置的方式进行合并
### pandas.merge 可根据一个或多个键将不同DataFrame中的行连接起来
### pandas.concat 可以沿着一条轴将多个对象堆叠在一起
### 实例方法combine_first 可以将重复数据编接在一起 用一个对象中的值充填另一个对象中的缺失值

### 数据库风格的DataFrame合并

#### 数据集的合并（merge）或连接（join）运算时通过一个或多个键将行链接起来的--pandas.merge（）函数

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                     'data2': range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # 多对一的合并 df1中的数据有多个标记为a和b的行 df2中key列的每个值仅仅对应一行
```

```python
pd.merge(df1, df2, on='key') # merge会将重叠的列的列名当作键
```

```python
# 两个对象的列名字不同，也可以分别进行指定
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 
                   'data1': range(7)})

df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer') # 默认下merge做的是inner 内链接交集 外连接取得是并集
```

```python
# 多对多连接产生的是行的笛卡儿积
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how='left') # 左边的DataFrame有3个’b'行 右边2个 最终结果有6行
```

```python
pd.merge(df1, df2, how='inner')
```

```python
# 需要根据多个键进行合并 传入一个由列名组成的列表即可
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3 ]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'], 
                      'rval': [4, 5, 6, 7]})
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
# 合并运算还需要考虑的最后一个问题是对重复列名的处理
pd.merge(left, right, on='key1')
```

```python
# suffixes选项 用于指定附加到左右两个DataFrame对象的重叠列名上的字符串
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

```python
# merge函数的参数
# left                参与合并的左侧DataFrame
# right               参与合并的右侧DataFrame
# how                 'inner' 'outer' 'left' 'right'
# on                  用于连接的列名
# left_on             左侧DataFrame中用作连接键的列
# right_on            右侧DataFrame中用作连接键的列
# left_index          将左侧的行索引用作其连接键
# right_index         类似于left_index
# sort                根据连接键对合并后的数据进行排序
# suffixes            字符串值元组 用于追加到重叠列名的末尾
# copy                设置为False   默认总是复制
```

### 索引上的合并

#### 当DataFrame中的连接键位于其索引中 可以传入left_index=True或者right_index=True(或两个都传)已说明索引应该被用作连接键

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
# 对于层次化索引 必须以列表形式指明合并键的多个列
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                      'key2': [2000, 2001, 2002, 2001, 2002], 
                      'data': np.arange(5)})
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
lefth
```

```python
righth
```

```python
# 必须以列表形式指明用作合并键的多个列
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')
```

```python
# 合并双方的索引也没问题
left2 = pd.DataFrame(np.arange(1, 7).reshape(3, 2), index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])

right2 = pd.DataFrame(np.arange(7, 15).reshape(4, 2), index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
left2
```

```python
right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left1.join(right1, on='key')
```

#### 对于Pandas对象，带有标签的轴能更近一步推广数组的连接运算
#### 具体考虑问题：
#### 如果各对象其他轴上的索引不同， 那些轴应该做并集还是交集
#### 如果各对象其他轴上的索引不同，那些轴应该是做并集还是交集
#### 结果对象中的分组需要各不相同吗
#### 用于连接的轴重要么
#### Pandas的contact函数提供了一种能够解决这些问题的可靠方法

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])

pd.concat([s1, s2, s3])
```

```python
# 默认情况下 concat是在axis=0上工作的 最终产生一个新的Series 如果传入axis=1 就会变成一个DataFrame
pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1 * 5, s3])
```

```python
s4
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner') # 交集
```

```python
# join_axes指定要在其他轴上使用的索引
pd.concat([s1, s4], axis=1, join_axes=[['a', 'c', 'b', 'e']])
```

```python
# 在连接轴上创建一个层次化索引
result = pd.concat([s1, s2, s3], keys=['one', 'two', 'three'])
result
```

```python
result.unstack()
```

```python
# 如果沿着aixs=1进行合并 keys会变成列头
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
# 同样的逻辑对于DataFrame对象也是一样的
df1 = pd.DataFrame(np.arange(6).reshape(3,2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                   columns=['three', 'four'])
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'])
```

```python
# 如果传入的不是列表而是一个字典 则字典的键就会被当作keys选项的值
pd.concat({'level1': df1, 'level2': df2}, axis=1)
```

```python
# 还有两个用于层次化索引的创建方式 names levels
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'],
          names=['upper', 'lower'])
```

```python
# 重新产生一组行索引
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
df1
```

```python
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
# concat函数的参数
# objs              参与连接的pandas对象的列表或字典 唯一必须的参数
# axis              指明连接的轴向
# join              'inner' 'outer' 默认为'outer'指明其他轴向上的索引是按交集还是并集进行合并的
# join_axes         指明用于其他n-1条轴的索引，不执行交集 并集运算
# keys              与连接对象有关 用于形成连接轴向上的层次化索引
# levels            用于指定层次化索引各级别上的索引 如果设置了keys的话
# names             用于创建分层级别的名称 如果设置了keys和levels的话
# verify_integrity  检查结果对象新轴上的重复情况 如果发现则引发异常   默认允许重复
# ignore_index      不保留连接轴上的索引 产生一组新的索引
```

### 离散化和面元划分

#### 连续数据常常被离散化或拆分为“面元”  使用pd.cut函数

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

```python
cats.value_counts()
```

```python
# 使用pd.cut函数返回一个特殊的Categorical对象
## 实际上含有表示不同分类名称的levels数组以及一个为年龄数据进行标号的labels属性
cats.labels
```

```python
cats.view
```

```python
cats.value_counts()
```

```python
# 和区间的数学符号一样 圆括号表示开端 方括号表示闭端
# 哪边是闭端可以使用right=False进行修改
pd.cut(ages, bins, right=False)
```

```python
# 设置分组名称 labels参数
groups_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
n = pd.cut(ages, bins, labels=groups_names)
```

```python
n.value_counts()
```

```python
# 如果向cut传入的是面元数量而不是确切的面元边界 
# 则会根据数据的最小值和最大值计算等长面元
data = np.random.rand(20)
data # 创建一个给定类型的数组 将其填充在一个均匀分布的随机样本[0, 1)中
```

```python
pd.cut(data, bins=4, precision=2) # precition 精度 保持小数点后两位
```

```python
# qcut是一个非常类似于cut的函数，可以根据样本分位数对数据进行面元划分
data = np.random.randn(1000) # 正态分布
cats = pd.qcut(data, 4) # 按四分位数进行切割
cats
```

```python
cats.value_counts()
```

```python
n = pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
```

```python
n.value_counts()
```

### 字符串对象方法

```python
val = 'a,b, guido'
```

```python
# string 转化为list
val.split(',')
```

```python
l = [x.strip() for x in val.split(',')]
l
```

```python
# 列表转字符串 join方法传入一个列表或者元组
','.join(l)
```
