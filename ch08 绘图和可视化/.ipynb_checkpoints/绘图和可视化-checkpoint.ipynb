### 注解以及在Subplot上绘图
#### 除了标准的图表对象 还可以绘制一些自定义的注解 比如文本， 箭头 或其他图形等
#### text（文本） arrow（箭头） annotate（注释）

#### 小知识点：numpy等差数列(numpy.linspace)和移动spines(标明数据区域）

```python
# 中文显示问题
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

fig = plt.figure(figsize=(8, 5), dpi=80)
plt.subplot(111)

# 创建正弦函数和余弦函数的图像 即 y=cosx y=sinx
## 创建等差数列  参数num 表示等差数列分成256分 endpoint=True 表示第257份为np.pi
x = np.linspace(-np.pi, np.pi, num=256, endpoint=True) # 等差数列 start(n1):-np.pi end(n_end):np.pi, 
s = np.sin(x) # y值
c = np.cos(x) # x值

plt.plot(x, s, 'b--', alpha=0.5, label='$y=\sin x$', linewidth=2.5)
plt.plot(x, c, 'r--', alpha=0.5, label='$y=\cos x$', linewidth=2.5)
plt.legend(loc='best')

# 移动spines 记录数据边界区域的线
ax = plt.gca() # stand for get current axis
ax.spines['right'].set_color('none') # 设置颜色('none')去除右边界
ax.spines['top'].set_color('none') # 设置颜色('none')去除上边界
ax.xaxis.set_ticks_position('bottom') # 设置下边界 为x轴刻度位置
ax.spines['bottom'].set_position(('data', 0)) # 设置下边界位置(('data', 0))
ax.yaxis.set_ticks_position('left') # 设置左边界 为y轴刻度位置
ax.spines['left'].set_position(('data', 0)) # 设置左边界位置(('data', 0))

# 设置x轴范围xlim 显示刻度xticks y轴范围ylim 显示刻度yticks
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', 0, r'$\pi/2$', r'$\pi$'], fontsize=12)

plt.ylim(s.min()*1.1, s.max()*1.1)
plt.yticks([-1, 0, 1],
           [r'$-1$', r'$0$', r'$1$'],fontsize=12)

plt.title('正弦函数 余弦函数', fontsize=16)
plt.show()


```

### matplotlib绘图注解的两种基本方式：plt.annotate plt.text

```python
# 绘制y=2x+1的函数图像

# 解决显示中文和负数问题
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

# 创建画板
plt.figure(dpi=150)
# 创建函数图形
# x值 y值
x = np.linspace(-3, 3, num=50)
y = 2 * x + 1

plt.plot(x, y, label=r'y=2x+1', color='yellow', linewidth=2.5, alpha=50)
plt.legend(loc='best')

# 移动spines 数据边界
ax = plt.gca() # stand for current axis
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# 设置x轴范围 刻度 y轴范围 刻度
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.xticks(np.arange(-3, 4), fontsize=10)

plt.ylim(y.min()*1.1, y.max()*1.1)
plt.xticks(np.arange(-3, 4), fontsize=10)

# 绘制注释点与虚线 (x0, y0)
x0 = 1
y0 = 2 * x0 + 1
plt.scatter(x0, y0, color='b', s=20)
plt.plot([x0, x0], [0, y0], color='b', linestyle='--', alpha=0.5)

# plt.annotate 对(x0, y0)这个点进行标注 注释
plt.annotate(r'$2x+1=%s$'% y0, xy=(x0, y0), # string is annotate content xy is annotate position
             xycoords='data', xytext=(+30, -30), # xycoords means base on xy to datemine annotate content position, xytext means annotate content position bias
             textcoords='offset points', fontsize=16, # textcoords content position 'offset point' not effect figure
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2")) # arrowprops set arrowstyle connectionstyle

# 添加注释text  plt.text
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 12, 'color': 'r'})
plt.title(r'函数$y=2x+1$及注释')
plt.show()
```

### 2008-2009年金融危机期间的重要日期

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import os

fig = plt.figure(dpi=150)
ax = fig.add_subplot(111)

data = pd.read_csv(r'C:\Users\Administrator\Desktop\计算机入门学习\支线：Python for Data Analysis\pydata-book-2nd-edition\pydata-book-2nd-edition\examples\spx.csv', encoding='utf-8', engine='python', index_col=0, parse_dates=True)
data.head()
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 50),
                xytext=(date, spx.asof(date) + 200),
                arrowprops=dict(facecolor='black'),
                horizontalalignment='left', verticalalignment='top')
    
# 设置坐标轴
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Import dates in 2008-2009 financial crisis')
```

```python
tips = pd.read_csv(r'C:\Users\Administrator\Desktop\计算机入门学习\支线：Python for Data Analysis\pydata-book-2nd-edition\pydata-book-2nd-edition\examples\tips.csv', encoding='utf-8', engine='python')
tips 
party_counts = pd.crosstab(index=tips['day'], columns=tips['size'])
party_counts

# 规格化 使得各行的和为1 
party_pcts = party_counts.div(party_counts.sum(1).astype(float), axis=0)
party_pcts

party_pcts.plot(kind='barh', stacked=True)
plt.title('每天各种聚会的规模')
```

### 散布图
#### scatter plot 是观察两个一维数据序列之间的关系的有效手段

```python
macro = pd.read_csv(r'C:\Users\Administrator\Desktop\计算机入门学习\支线：Python for Data Analysis\pydata-book-2nd-edition\pydata-book-2nd-edition\examples\macrodata.csv', encoding='utf-8', engine='python')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]

trans_data = np.log(data).diff().dropna() # np.diff() 计算离差 out[n] = a[n+1] - a[n]
trans_data.head()
```

```python
# numpy.diff(a, n=1, axis=-1)
#Calculate the n-th discrete difference along the given axis.
#The first difference is given by out[n] = a[n+1] - a[n] along the given axis, higher differences are calculated by using diff recursively.
```

```python
plt.scatter(trans_data['m1'], trans_data['unemp'])
plt.title('Changes in log %s vs. log %s' % ('m1', 'unemp'))

pd.scatter_matrix(trans_data, diagonal='kde', color='k', alpha=0.3) # 散步图矩阵
```
